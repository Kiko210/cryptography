# General plot.**py**

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```
x = np.random.random((10, 1))  # 10 different numbers generated between 0 to 1 [random data]
print(x)
```

```
plt.plot(x, '*-')
plt.show()
```

```
x = np.linspace(0, 100, 50)  # generating 50 numbbers between 0 to 100, linspace gives float number.
y = np.power(x, .5)          # for each value of x, x power 0.5 = y
plt.plot(x, y, '+-')
plt.show()

sns.set()
sns.lineplot(x = x, y = y)
plt.show()
```

# **Plotting Data from Dataset**

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```
data = pd.read_csv('machine_learning/Plotting/mobile_cleaned.csv')
data.head()
```

```
ax = sns.scatterplot(x="stand_by_time", y="battery_capacity", data=data)
plt.show()
```

```
ax = sns.scatterplot(x = "stand_by_time", y = "battery_capacity", hue="thickness", data=data) # learn from seaborn.scatterplot
plt.show()
```

```
ax = sns.distplot(data["stand_by_time"]) # distances between data points
plt.show()
```

```
ax = sns.boxplot(x="is_liked", y="battery_capacity", data=data) # most preferred range (2400 - 3100) but it ranges from (1600-4100) with some outliers in 5000 region.
plt.show()
```

```
ax = sns.boxplot(x = "expandable_memory", y = "price", data=data)
plt.show()
```

```
# creating the bar plot
# creating the dataset
data = {'Audi':20, 'Mercedes':15, 'Lamborgini':30,
        'Mini cooper':35, 'Bentley': 58, "Rolls Royce":3}
carNames = list(data.keys())
counts = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(carNames, counts, color ='Green',
        width = 0.4)


plt.xlabel("Luxury cars")
plt.ylabel("No. of people holds it")
plt.title("Chennai based luxury car holders")
plt.show()
```
